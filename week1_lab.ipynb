```python
#Problem 2

prompts = [
    "Explain␣transformers?", "hi","please␣summarize␣this␣paragraph␣about␣GANs", "Generate␣an␣image_of_a_cat"
]

def is_good_prompt(prompt):
    length_condition = len(prompt) >= 20
    keyword_condition = '?' in prompt or 'please' in prompt.lower()
    return length_condition and keyword_condition

for prompt in prompts:
    result = is_good_prompt(prompt)
    if result:
        print(f"{prompt} -> {result}")
```

```python
#Problem 6

import math

a = [1, 2, 3]
b = [2, 4, 1]

def dot_product(a, b):
  return sum(x * y for x, y in zip(a, b))

def magnitude(i):
  return math.sqrt(sum(x**2 for x in i))

def cosine_similarity(a, b):
  return dot_product(a, b) / (magnitude(a) * magnitude(b))
dot_1 = dot_product(a, b)
cos_1 = cosine_similarity(a, b)

print(f"Dot Product: {dot_1}")
print(f"Cosine Similarity: {cos_1}")
```
