```python
#Problem 1

token=[18, 42, 37, 10, 55, 26]

total=sum(token)
aver=total/len(token)
min_val = min(token)
max_val = max(token)

print(f"total: {total}")
print(f"aver: {aver}")
print(f"min: {min_val}")
print(f"max: {max_val}")
```

```python
#Problem 2

prompts = [
    "Explain␣transformers?", "hi","please␣summarize␣this␣paragraph␣about␣GANs", "Generate␣an␣image_of_a_cat"
]

def is_good_prompt(prompt):
    length_condition = len(prompt) >= 20
    keyword_condition = '?' in prompt or 'please' in prompt.lower()
    return length_condition and keyword_condition

for prompt in prompts:
    result = is_good_prompt(prompt)
    if result:
        print(f"{prompt} -> {result}")
```
